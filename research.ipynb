```python
# libraries that work with nltk
str_data = ' '.join(data['reviews'])

sentences = nltk.sent_tokenize(str_data)
lemma = WordNetLemmatizer()


# cleaning the paragraph
import re
def cleaned_paragraph(sentences):  
    corpus = []
    for i in range(0, len(sentences)):
        text = re.sub('[^a-zA-Z]', ' ', sentences[i])
        text = text.lower()
        text = text.split()
        text = [lemma.lemmatize(x) for x in text if x not in set(stopwords.words('english'))]
        text = ' '.join(text)
        corpus.append(text)
    return corpus

corpus = cleaned_paragraph(sentences)
```

1. Vader Sentiment Scoring

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
from tqdm.notebook import tqdm

sid = SentimentIntensityAnalyzer()
```

```python
def polarity_score(sentences):
    for sentence in sentences:
         sid = SentimentIntensityAnalyzer()
         print(sentence)       
         ss = sid.polarity_scores(sentence)
         for k in sorted(ss):                            
            print('{0}: {1}, '.format(k, ss[k]), end='')        
         print()
ps = polarity_score(corpus)
ps
```
